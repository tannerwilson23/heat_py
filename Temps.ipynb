```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 0.25
kappa = 2.28E-5
L = 1
N = 20
dx = L/N
dt = alpha*dx**2/kappa

x_vals = np.zeros(21)
for i in range(len(x_vals)):
    x_vals[i] = i * dx
    
    
Temps = np.ndarray((100,N+1))

Temps[0] = 100*np.sin(x_vals*np.pi/L)
```

```python
plt.plot(x_vals,Temps[0])
```

```python
for i in range(1,Temps.shape[0]):
    for j in range(1,Temps.shape[1]-1):
        Temps[i,j] = alpha*Temps[i-1,j+1] + (1-2*alpha)*Temps[i-1,j] + alpha*Temps[i-1,j-1]
    Temps[i,0] = 0
    Temps[i,-1] = 0
```

```python
plt.plot(x_vals,Temps.T)
```
